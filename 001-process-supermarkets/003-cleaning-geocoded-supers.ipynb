### Merge supers and cities based on city name

```python
import Levenshtein
```

```python
supers_yango['1'] = 1
cities_gdf_supers['1'] = 1
```

```python
gdf_cross_merge = supers_yango.merge(cities_gdf_supers, on='1', suffixes=('',"_city"))
```

```python
gdf_cross_merge['city_similarity_score'] = gdf_cross_merge[['city','Muni_Heb']].apply(lambda x: Levenshtein.ratio(x['city'],x['Muni_Heb']), axis=1)
```

```python
gdf_cross_merge[gdf_cross_merge['city_similarity_score']>0.7]['Muni_Eng'].unique()
```

```python
gdf_cross_merge[gdf_cross_merge['city_similarity_score']>0.8]['Muni_Eng'].unique()
```

No shops in kfar kasem. This is a different place

```python
gdf_cross_merge[(gdf_cross_merge['city_similarity_score']>0.7)&(gdf_cross_merge['Muni_Eng']=='Kefar Kasem')]
```

```python
gdf_cross_merge = gdf_cross_merge[gdf_cross_merge['city_similarity_score']>0.8]
```

```python
gdf_cross_merge.explore()
```
