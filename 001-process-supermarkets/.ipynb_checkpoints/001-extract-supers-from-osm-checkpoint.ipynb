```python
import pandas as pd
import geopandas as gpd
```

```python
gdf = gpd.read_file("../data/osm_pois_israel.geojson")
```

```python
gdf.fclass.value_counts().head(50)
```

```python
shops = gdf[gdf.fclass.isin(['supermarket','convenience', 'kiosk'])]
```

```python
shops.shape
```

```python
# shops.explore()
```

## District geometry to remove shops in Palestine

```python
borders = gpd.read_file("../data/districts_borders.geojson")
```

```python
borders.crs
```

```python
borders.iloc[0].geometry
```

```python
shops_isr= gpd.sjoin(shops, borders, predicate='within', how='inner')
```

```python
shops_isr.explore()
```

```python
shops_isr.name.value_counts().head(20)
```

```python
shops_isr[shops_isr.fclass=='convenience'].name.value_counts()
```

```python
shops_isr.to_file("../data/shops_israel_osm.geojson")
```
